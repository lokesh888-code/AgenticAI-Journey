Sync Example


```python
import time

def fetch_weather_sync(city):
    print(f"Fetching weather for {city}...")
    time.sleep(2)  # Simulate a delay in fetching data
    return f"Weather data for {city}: Sunny, 25°C"

cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]

start_time = time.time()


for city in cities:
    result = fetch_weather_sync(city)
    print(result)

print(f"Total time taken: {time.time() - start_time:.2f} seconds")
```

```python
end_time = time.time()
print("start_time:", start_time)
print("end_time:", end_time)
print("elapsed:", end_time - start_time)
```

Sequential Main

```python
import time
import asyncio
async def api_call(name):
    print(f"{name} Started.")
    await asyncio.sleep(2)
    print(f"{name} Finished.")
    return name
async def sequential_calls():
    start_time = time.time()
    result1 = await api_call("Call 1")
    result2 = await api_call("Call 2")
    result3 = await api_call("Call 3")
    print(result1, result2, result3)
    print("Total",round(time.time() - start_time, 2), "seconds")

await sequential_calls()
```

Gather Main

```python
async def api_call(name):
    print(f"{name} Started.")
    await asyncio.sleep(2)
    print(f"{name} Finished.")
    return name

async def concurrent_calls():
    start_time = time.time()
    results = await asyncio.gather(
        api_call("Call 1"),
        api_call("Call 2"),
        api_call("Call 3")
    )
    print(results)
    print("Total", round(time.time() - start_time, 2), "seconds")

await concurrent_calls()
```

Create_task

```python
async def fetch_date(id,sleep_time):
   print(f"Coroutine {id} starting to fetch data...")

   await asyncio.sleep(sleep_time)  # Simulate a delay in fetching data
   return{
      "id": id,
      "data" : f"Data fetched after {sleep_time} seconds"
   }

async def main():
   start_time = time.time()
   task1 = asyncio.create_task(fetch_date(1, 2))
   task2 = asyncio.create_task(fetch_date(2, 3))
   task3 = asyncio.create_task(fetch_date(3, 1))
   
   result1 = await task1
   print(result1)
   
   result3 = await task3
   print(result3)

   result2 = await task2
   print(result2)
   print("Total", round(time.time() - start_time, 2), "seconds")

await main()
```
